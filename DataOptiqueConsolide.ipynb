```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
MCD14ML=pd.read_csv('/content/drive/MyDrive/PFE_CIRAD_AMAP/1_MCD14ML/Outputs_ThermalAnomalies/MCD14ML_TA.csv')
#area in km2
Grassland=pd.read_csv('/content/drive/MyDrive/PFE_CIRAD_AMAP/1_MCD14ML/Inter_data_ThermalAnomalies/Grassland_Area_part_all2.csv')
Cropland=pd.read_csv('/content/drive/MyDrive/PFE_CIRAD_AMAP/1_MCD14ML/Inter_data_ThermalAnomalies/Cropland_Area_part_all2.csv')
TreeCover=pd.read_csv('/content/drive/MyDrive/PFE_CIRAD_AMAP/1_MCD14ML/Inter_data_ThermalAnomalies/TreeCover_Area_part_all2.csv')
Shrubland=pd.read_csv('/content/drive/MyDrive/PFE_CIRAD_AMAP/1_MCD14ML/Inter_data_ThermalAnomalies/Shrubland_Area_part_all2.csv')
```

```python
MCD14ML_data=MCD14ML.groupby(['Pays','WordlCover'])['FRP'].count().reset_index()
MCD14ML_data.columns=['Pays','WordlCover','Fréquence']

Grassland["WordlCover"]="Grassland"
Grassland.columns=['Pays','Area(km2)','WordlCover']

Cropland["WordlCover"]="Cropland"
Cropland.columns=['Pays','Area(km2)','WordlCover']

TreeCover["WordlCover"]="Tree cover"
TreeCover.columns=['Pays','Area(km2)','WordlCover']

Shrubland["WordlCover"]="Shrubland"
Shrubland.columns=['Pays','Area(km2)','WordlCover']

Concat=pd.concat([Grassland,Cropland,TreeCover,Shrubland],axis=0,ignore_index=True)

Concat.sample(n=15,random_state=15)
```

```python
result = pd.merge(MCD14ML_data, Concat, on=["Pays","WordlCover"])
result.sample(n=15,random_state=15)
```

```python
result["Numero/km2"]=result["Fréquence"]/result["Area(km2)"]
result.sample(n=15,random_state=15)
```

```python
ax=sns.barplot(result, x="Numero/km2", y="Pays", hue="WordlCover")
ax.set_title('Thermal Anomalies Andes Tropicales entre 2000-2024')
```
